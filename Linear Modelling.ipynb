### Function to calculate VIF(Variance Inflation Factor):

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
## Library required to calculate VIF
```

```python
def calVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
## This function will give a pandas dataframe with the VIF values of each column
```

# Step 8: Model Evaluation

**Calculating  R-squared score:**

```python
from sklearn.metrics import r2_score
## Library required to calculate r2 value.
```

```python
r_squared = r2_score(y_test, y_test_pred)
r_squared
## Calculating R-squared score for test data
```

### Summarising the R-Square values for Test and Train data sets:

```python
r_squared_train=r2_score(y_train, y_train_pred)
r_squared_test = r2_score(y_test, y_test_pred)
## Calculating the r square for train and test data 
```

```python
AdjR_squared_train=lm_6.rsquared_adj
## Calculating the adj r square for train data 
```

**We know that, Adjusted R2 = 1  [(1-R2)*(n-1)/(n-k-1)]**

```python
AdjR_squared_test=1-(((1-r_squared_test)*(X_test.shape[0]-1))/(X_test.shape[0]-X_test.shape[1]-1))
```

```python
print("Train data set R-square :",round(r_squared_train,4))
print("Test data set R-square :",round(r_squared_test,4))
print("Train data set Adjusted R-square:",round(AdjR_squared_train,4))
print("Test data set Adjusted R-square:",round(AdjR_squared_test,4))
```
